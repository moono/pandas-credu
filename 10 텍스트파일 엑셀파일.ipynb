```python
def 데이터프레임추출(시트프레임):
    색인시작목록 = pd.Series(시트프레임.index[시트프레임.notnull().sum(1) == 1])
    색인종료목록 = 색인시작목록[1:]
    색인종료목록 = 색인종료목록.append(pd.Series(len(시트프레임)))
    색인범위목록 = list(zip(색인시작목록, 색인종료목록))
    
    프레임목록 = []
    for 범위 in 색인범위목록:
        # 범위가 2보다 작으면 건너뛴다
        if 범위[1] - 범위[0] < 2:
            continue
            
        시작, 종료 = 범위
        프레임 = 시트프레임.iloc[시작:종료]
        프레임 = 프레임.reset_index(drop=True)
        프레임목록.append(프레임)
        
    return 프레임목록
```

```python
프레임목록 = 데이터프레임추출(국제0803)
for 표 in 프레임목록:
    print(표)
```
