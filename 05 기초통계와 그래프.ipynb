```python
게임사용자 = pd.read_excel('data/game_users.xlsx', index_col='장르')
게임사용자
```

```python
# pie graph 사용시 subplots 인자를 꼭 사용해야 함
게임사용자['누적사용자수'].plot(kind='pie', subplots=True)
```

```python
게임사용자['만족도'].plot(kind='pie', subplots=True)
```

```python
게임사용자['만족도'].plot(kind='bar')
```

```python
게임사용자['만족도'].plot(kind='barh')
```

```python
그래프 = 게임사용자['만족도'].plot(kind='barh')
그래프.set_xticklabels('{:3.1%}'.format(x) for x in 그래프.get_xticks())
```

```python
만족도수 = 게임사용자['누적사용자수'] * 게임사용자['만족도']
만족도수.plot(kind='barh')
```
