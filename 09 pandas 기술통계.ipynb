```python
import pandas as pd
```

```python
NC2015 = pd.read_excel('./data/NC Dinos 2015.xlsx')
NC2015 = NC2015[['선수명', '경기', '타석', '안타', '홈런']]
NC2015 = NC2015.set_index('선수명')
NC2015[:3]
```

```python
NC시즌 = []
for 연도 in range(2013, 2016):
    파일경로 = './data/NC Dinos {:d}.xlsx'.format(연도)
    프레임 = pd.read_excel(파일경로)
    프레임 = 프레임[['선수명', '경기', '타석', '안타', '홈런']]
    프레임['시즌'] = 연도
    프레임 = 프레임.set_index(['선수명', '시즌'])
    NC시즌.append(프레임)
```

```python
NC15 = NC시즌[-1]
NC15[:3]
```
