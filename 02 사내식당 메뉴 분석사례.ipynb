```python
열목록 = ['dine_type', 'corner', 'menu', 'use_count']
식당 = 식당[열목록]
식당[:3]
```

```python
열제목 = {'dine_type': '시간대', 'corner': '코너', 'menu': '메뉴', 'use_count': '주문수량'}
식당 = 식당.rename(columns=열제목)
식당[:3]
```

```python
시간대 = {'breakfast': '아침', 'lunch': '점심', 'dinner': '저녁'}
식당['시간대'] = 식당['시간대'].replace(시간대)
식당[:3]
```

```python
restaurant = 식당
```

```python
pivot_01 = restaurant.pivot_table('주문수량', aggfunc='mean', index='코너', columns='시간대')
pivot_01
```

```python
time_line = ['아침', '점심', '저녁']
pivot_01 = pivot_01[time_line]
pivot_01
```

```python
pivot_01.sum()
```

```python
sum_col = pivot_01.sum(1)
```

```python
pivot_01['합계'] = sum_col
pivot_01
```

```python
pivot_01.sort_values(by='합계')
```

```python
pivot_01.sort_values(by='합계', ascending=False)
```

```python
pivot_01[time_line].plot(kind='bar')
```

```python
pivot_01[time_line].plot(kind='barh')
```

```python
pivot_01[time_line].plot(kind='barh', stacked=True)
```

```python
pivot_01.to_excel('코너별 시간대별 주문수량.xlsx')
```
