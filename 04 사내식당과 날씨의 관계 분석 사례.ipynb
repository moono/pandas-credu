```python
import pandas as pd
```

```python
날씨 = pd.read_excel('data/weather.xlsx')
날씨[:5]
```

```python
식당 = pd.read_excel('./data/food_order.xlsx')
식당[:5]
```

# 날씨와 식당 자료에서 필요한 열만 선택

```python
날씨 = 날씨[['date', 'location', 'avgTemp', 'weather']]
날씨[:5]
```

```python
식당 = 식당[['date', 'corner', 'menu', 'use_count']]
식당[:5]
```

# 영문 열을 한국어로 변경

```python
날씨 = 날씨.rename(columns={'date': '날짜', 'location': '지역', 'avgTemp': '일평균', 'weather': '기상'})
날씨[:3]
```

```python
식당 = 식당.rename(columns={'date': '날짜', 'corner': '코너', 'menu': '메뉴', 'use_count': '주문수량'})
식당[:3]
```

# 값(value) 자체를 변경 가능

```python
날씨 = 날씨.replace({'seoul': '서울', 'suwon': '수원'})
날씨[:5]
```

# 정리된 자료를 엑셀 파일로 저장

```python
날씨.to_excel('날씨.xlsx')
식당.to_excel('식당.xlsx')
```

# 다시 저장된 파일 불러오기

```python
날씨 = pd.read_excel('날씨.xlsx')
날씨[:5]
```

```python
식당 = pd.read_excel('식당.xlsx')
식당[:5]
```

# 원하는 열의 종류와 개수를 미리보기

```python
날씨['지역'].value_counts()
```

```python
수원날씨 = 날씨[날씨.지역 == '수원']
수원날씨[:3]
```

# 데이터 프레임 합치기

```python
식당날씨 = pd.merge(식당, 수원날씨, on='날짜')
식당날씨[:5]
```

# 날짜 열을 색인으로 설정

```python
식당날씨 = 식당날씨.set_index('날짜')
식당날씨[:5]
```

# 기상에 따른 주문수량 확인하기

```python
식당날씨.pivot_table('주문수량', aggfunc='mean', index='코너', columns='기상')
```

```python
식당날씨['기상'].value_counts()
```

```python
눈비여부 = 식당날씨['기상'].str.contains('눈|비')
눈비여부[:5]
```

# 열 단위 문자 처리 - series

```python
코너별_눈비_주문평균 = 식당날씨.pivot_table('주문수량', index='코너', columns=눈비여부)
코너별_눈비_주문평균
```

```python
코너별_눈비_주문평균 = 코너별_눈비_주문평균.rename(columns={
    True: '눈비',
    False: '눈비 안 옴'
})
코너별_눈비_주문평균
```

```python
코너별_눈비_주문평균 = 코너별_눈비_주문평균.sort_values(by='눈비', ascending=False)
코너별_눈비_주문평균
```

```python
코너별_눈비_주문평균.plot(kind='barh')
```

```python
코너별_눈비_주문평균.plot(kind='barh', stacked=True)
```

```python
합계 = 코너별_눈비_주문평균.sum(axis=1)
합계
```

```python
코너별_눈비_주문비율 = 코너별_눈비_주문평균.div(합계, axis=0)
코너별_눈비_주문비율
```

```python
코너별_눈비_주문비율.plot(kind='barh', stacked=True)
```

```python
눈비_평균차 = 코너별_눈비_주문비율['눈비'] - 코너별_눈비_주문비율['눈비'].mean()
눈비_평균차
```

```python
눈비_평균차.sort_values().plot(kind='barh')
```
